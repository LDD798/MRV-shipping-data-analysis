## How many miles did each vessel in the database travel?

Distance = Fuel consumption / (average Fuel consumption per n mile)

```python
# Show division by zero strings
for i in range(len(df_sample)):
    item = df_sample['Annual average Fuel consumption per distance [kg / n mile]'].iloc[i]
    if not utils.is_float(item):
        print(item)
```

```python
# Convert strings to nans
average_fuel_consuption = df_sample['Annual average Fuel consumption per distance [kg / n mile]'].copy()
print(average_fuel_consuption.isna().mean())
average_fuel_consuption = pd.to_numeric(average_fuel_consuption, errors='coerce')
print(average_fuel_consuption.isna().mean())
```

```python
# Convert m tonnes to kg
df_sample['Total fuel consumption [kg]'] = df_sample['Total fuel consumption [m tonnes]'] * 1000
```

```python
distance = df_sample['Total fuel consumption [kg]'] / average_fuel_consuption
print(f"Median: {distance.median()}")
print(f"Mean: {distance.mean()}")
```

```python
df_sample['Distance'] = distance
fig = px.box(df_sample, y='Ship type', x='Distance', points=False, color='Ship type', title='Total Distance Travelled [n miles]')#, width=1000, height=1000)
fig.show()
```
